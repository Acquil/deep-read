```
import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from os import listdir,rename
```

```
model = tf.keras.models.load_model("/content/model_v1_renet50")
```

```
!mkdir slides
!mkdir others
```

```
!unzip '/content/deep-read/data/training/CNN/video conference screens/images_extracted_from_google_meet_recording.zip'
```

```
folder = '/content/images_extracted_from_google_meet_recording/'
for img_path in listdir(folder):
  img = load_img(folder+img_path, target_size=(299, 299))
  x = img_to_array(img)  # Numpy array with shape (150, 150, 3)
  x = x.reshape((1,) + x.shape)  # Numpy array with shape (1, 150, 150, 3)

  # Rescale by 1/255
  x /= 255

  class_ = model.predict(x)
  if class_[0][0] >0.5:
    rename(folder+img_path,'/content/slides/'+img_path)
  else:
    rename(folder+img_path,'/content/others/'+img_path)
```
